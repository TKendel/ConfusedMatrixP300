```python
from scipy.io import loadmat
import pandas as pd

file_paths = [
    './p300/S1.mat',
    './p300/S2.mat',
    './p300/S3.mat',
    './p300/S4.mat',
    './p300/S5.mat'
]

dfs = []

for i, file_path in enumerate(file_paths):
    data = loadmat(file_path)
    eeg_data = data['y']
    trigger_data = data['trig']
    
    df = pd.DataFrame(eeg_data, columns=[f"Ch_{i+1}" for i in range(eeg_data.shape[1])])
    df['trigger'] = trigger_data
    df['subject'] = f"S{i+1}"
    
    dfs.append(df)

eeg_df = pd.concat(dfs, ignore_index=True)
eeg_df.to_csv('eeg_data_output.csv', index=False)
```
